```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('data/president_county_candidate.csv')
results
```

```python
all_candidates = pd.DataFrame(results['candidate'].value_counts())
vote_count = results[["candidate", "total_votes", 'party']]
vote_count = vote_count.groupby(['candidate'])

county_count = vote_count["candidate"].count()
votes = vote_count['total_votes'].sum()

top_votes = pd.DataFrame({"On Ballot" : county_count,
                         "Total Votes" : votes})

top_votes = top_votes.sort_values(["On Ballot", "Total Votes"],  ascending=[False,False])
top_votes = top_votes.iloc[2:]
top_votes
```

```python
plt.scatter((top_votes['On Ballot']), (top_votes['Total Votes']), marker='o', facecolors="blue", edgecolors="black", alpha=0.75)
```
